````python
print(f"Variables en el censo: {meta.column_names_to_labels}")
print(f"Variables en la ENH 09:{metaenh09.column_names_to_labels}")
print(f"Variables en la ENH 11:{metaenh11.column_names_to_labels}")
````

````python
print("Iniciando creación de variables homologadas...")
inicio = time.time()
#sexo
censo2010['sexo'] = censo2010['P27_SEXO'].map(etiqueta_sexo)
enhogar0910['sexo'] = enhogar0910['HL4'].map(etiqueta_sexo)
enhogar11['sexo'] = enhogar11['H202'].map(etiqueta_sexo)

#zona
censo2010['zona'] = censo2010['ZONA'].map(etiqueta_zona)
enhogar0910['zona'] = enhogar0910['HH6'].map(etiqueta_zona)
enhogar11['zona'] = enhogar11['HZONA'].map(etiqueta_zona)

#lectoescritura
censo2010['lectoescritura'] = censo2010['P35_SABE_LEER'].map(etiqueta_lectoescritura)
#la enh 2009 no trajo lectoescritura
enhogar11['lectoescritura'] = enhogar11['H212'].map(etiqueta_lectoescritura)

#region
censo2010['region_clean'] = censo2010['REGION'].map(etiqueta_region)
enhogar0910['region_clean'] = enhogar0910['Region'].map(etiqueta_region)
enhogar11['region_clean'] = enhogar11['REGION'].map(etiqueta_region)

#provincia
censo2010['provincia_clean'] = censo2010['PROVINCIA'].map(etiqueta_provincia)
enhogar0910['provincia_clean'] = enhogar0910['HH7'].map(etiqueta_provincia)
enhogar11['provincia_clean'] = enhogar11['HPROVI'].map(etiqueta_provincia)
final = time.time()
print(f"Proceso terminado en {(final - inicio):.2f} segundos")
````

#### El estado conyugal

Según el listin diario (https://www.diariolibre.com/actualidad/en-los-ltimos-cinco-aos-los-divorcios-en-la-repblica-dominicana-promedian-ms-de-17-mil-LNDL416508), para el año 2011 se registró un total de 17,181 parejas divorciadas. Suponiendo un comportamiento lineal durante el año, habría aproximadamente 1,431.75 divorcios por mes, lo que supone que, al 20 de septiembre, donde inició el levantamiento de la ENHOGAR 2011, se habían divorciado aproximadamente 13, 840.2 personas. Esto solo en este estado conyugal. No sabemos los cambios surgidos en las otras categorías.

Por esta razón, es menester evaluar profundamente si se podrá tomar el estado conyugal como una variable confiable para el linkage.

`Nota: Debido a que las fuentes levantan información de forma diferente, será necesario hacer la creación del estado conyugal a mano.`


````python
print(f"La variable H208, cuya forma de preguntar fue '{metaenh11.column_names_to_labels['H208']}', tiene las categorías:")  #ver label de la variable
print(metaenh11.value_labels[metaenh11.variable_to_label['H208']])
````

````python
# estado conyugal
print(f"Estado conyugal en la ENH 2011: {enhogar11['H208'].value_counts().sort_index()}")  #sort_index() muestra el orden de la variable, no de los pesos de la frecuencia.
print("="*50)
print("La ENHOGAR 200 no trajo la variable sobre estado conyugal")
print("="*50)

enhogar11['estado_conyugal'] = enhogar11['H208'].map(
    {
    1: 'casado',
    2: 'unido',
    3: 'viudo',
    4: 'divorciado',
    5: 'separado de un matrimonio',
    6: 'separado de una union libre',
    7: 'soltero',
    9: 'SinInfo'
}
)

print(f"Estado conyugal : {enhogar11['estado_conyugal'].value_counts()}")

````

````python

print(f"La variable P55_ESTADO_CIVIL, cuya forma de preguntar fue '{meta.column_names_to_labels['P55_ESTADO_CIVIL']}', tiene las categorías:")  #ver label de la variable
est_con_censo = meta.variable_to_label['P55_ESTADO_CIVIL']

print(meta.value_labels[est_con_censo])

censo2010['estado_conyugal'] = censo2010['P55_ESTADO_CIVIL'].map(
    {
    1: 'separado de un matrimonio',
    2: 'divorciado',
    3: 'viudo',
    4: 'separado de una union libre',
    5: 'casado',
    6: 'unido',
    7: 'soltero',
    9: 'SinInfo'
    }
)

print("="*50)
print(f"Estado conyugal en el censo: {censo2010['estado_conyugal'].value_counts()}")
````
